```python
import glob
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
```

```python
def colorbar_array(x):
    y = xr.open_dataset(x)
    y = y.Qs_summed
    y = np.flipud(y) ### <-- empirically derived
    return y
```

```python
def Qs_subplot(x):
    y = xr.open_dataset(x)
    y = y.Qs_summed
    y = np.flipud(y)
    return y
def Qsb_subplot(x):
    y = xr.open_dataset(x)
    y = y.Qsb_summed
    y = np.flipud(y)
    return y

def Qs_subplot_pcolormesh(x):
    y = xr.open_dataset(x)
    y = y.Qs_summed
    return y
def Qsb_subplot_pcolormesh(x):
    y = xr.open_dataset(x)
    y = y.Qsb_summed
    return y
```

```python
x = glob.glob('clipped__co/*.nc')
x1 = xr.open_dataset(x[1])
x1
```

```python
a0 = sorted(glob.glob('clipped__br/*.nc'))
a1 = sorted(glob.glob('clipped__co/*.nc'))
a2 = sorted(glob.glob('clipped__ct/*.nc'))
a3 = sorted(glob.glob('clipped__ms/*.nc'))

# print(len(a0),len(a1), len(a2), len(a3))

```

```python
import matplotlib as mpl
import matplotlib.gridspec as gridspec

cmap = mpl.cm.cividis_r
norm = mpl.colors.Normalize(vmin=5, vmax=10)
fontsize=20

plt.rcParams["font.family"] = "DejaVu Serif"
plt.rcParams['figure.dpi'] = 600
plt.rcParams['axes.linewidth'] = 2
plt.rcParams["font.weight"] = "bold"
fig = plt.figure(figsize=(20,10))


gs = gridspec.GridSpec(3,4,height_ratios=[1,1,.05],width_ratios=[1,1,1,1])

doy = 156+366+365+365+365+366+365 ###controls which number in list of .nc files starting with january 1 2015
print(x[doy])

ax0 = fig.add_subplot(gs[0, 0])
plt.imshow(Qs_subplot(a0[doy]),cmap=cmap,vmin=0,vmax=5,aspect='auto')
plt.title('Bear River',fontsize=fontsize,fontweight='bold')
plt.ylabel('Surface \nFlow',rotation=0, labelpad=50,fontsize=fontsize,fontweight='bold'); plt.xlabel(''); 
plt.xticks(fontsize=fontsize); plt.yticks(fontsize=fontsize)

ax4 = fig.add_subplot(gs[1, 0])
plt.imshow(Qsb_subplot(a0[doy]),cmap=cmap,vmin=0,vmax=5,aspect='auto')
plt.ylabel('Subsurface \nFlow',rotation=0, labelpad=50,fontsize=fontsize,fontweight='bold'); plt.xlabel(''); 
plt.xticks(fontsize=fontsize); plt.yticks(fontsize=fontsize)

ax1 = fig.add_subplot(gs[0, 1])
plt.imshow(Qs_subplot(a1[doy]),cmap=cmap,vmin=0,vmax=5,aspect='auto')
plt.title('Colorado',fontsize=fontsize,fontweight='bold')
plt.xticks(fontsize=fontsize); plt.yticks(fontsize=fontsize)


ax5 = fig.add_subplot(gs[1, 1])
plt.imshow(Qsb_subplot(a1[doy]),cmap=cmap,vmin=0,vmax=5,aspect='auto')
plt.xticks(fontsize=fontsize); plt.yticks(fontsize=fontsize)

ax2 = fig.add_subplot(gs[0, 2])
plt.imshow(Qs_subplot(a2[doy]),cmap=cmap,vmin=0,vmax=5,aspect='auto')
plt.title('Connecticut',fontsize=fontsize,fontweight='bold')
plt.xticks(fontsize=fontsize); plt.yticks(fontsize=fontsize)

ax6 = fig.add_subplot(gs[1, 2])
plt.imshow(Qsb_subplot(a2[doy]),cmap=cmap,vmin=0,vmax=5,aspect='auto')
plt.xticks(fontsize=fontsize); plt.yticks(fontsize=fontsize)

ax3 = fig.add_subplot(gs[0, 3])
plt.imshow(Qs_subplot(a3[doy]),cmap=cmap,vmin=0,vmax=5,aspect='auto')
plt.title('Mississippi',fontsize=fontsize,fontweight='bold')
plt.xticks(fontsize=fontsize); plt.yticks(fontsize=fontsize)

ax7 = fig.add_subplot(gs[1, 3])
plt7 = ax7.imshow(Qsb_subplot(a3[doy]),cmap=cmap,vmin=0,vmax=5,aspect='auto')
plt.xticks(fontsize=fontsize); plt.yticks(fontsize=fontsize)

cbar = plt.colorbar(plt7,cax=plt.subplot(gs[2,1:3]),orientation='horizontal')
cbar.ax.set_title('kg/$\mathregular{m^2}$',fontsize=fontsize,y=-4,fontweight='bold')
cbar.ax.tick_params(labelsize=fontsize) 

plt.tight_layout()
# plt.suptitle('yo',y=.91,x=.47,fontsize=24)
plt.savefig('../files_for_paper/fig2.png',bbox_inches='tight')
```
